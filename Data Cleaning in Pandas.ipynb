```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\mfuazi88\Desktop\DATA Analyst\Pandas\Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates()
```

```python
df=df.drop(columns='Not_Useful_Column')
```

```python
df["Last_Name"]=df["Last_Name"].str.lstrip("...")
df["Last_Name"]=df["Last_Name"].str.lstrip("/")
df["Last_Name"]=df["Last_Name"].str.rstrip("_")
df
```

```python
df["Phone_Number"]=df["Phone_Number"].astype(str).replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:11])
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
```

```python
df
```

```python
df[['Street Address','State','Zip Code']]=df['Address'].str.split(',', expand=True)
df
```

```python
df['Paying Customer']=df['Paying Customer'].astype(str).replace('Yes','Y')
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df
```

```python
df=df.drop(columns='Not_Useful_Column')
df
```

```python
df

```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].astype(str).replace('Yes','Y')
df
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x, inplace=True)
df
        
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x, inplace=True)
df
        
```
